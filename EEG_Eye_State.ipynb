```python
import pandas as pd
from scipy.io import arff
import numpy as np
import matplotlib.pyplot as plt
data = arff.loadarff('EEG Eye State.arff')
train= pd.DataFrame(data[0])
train.head(6656)
```

```python
X = train.iloc[:, :-1].values
y = train.iloc[:, -1].values
```

```python
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:-1])
X[:, 1:-1] = imputer.transform(X[:, 1:-1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

KNN Hiperparametre degişiklikleri

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Model ve parametreler
classifier = KNeighborsClassifier()
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'metric': ['minkowski', 'euclidean', 'manhattan'],
    'p': [1, 2]
}

# GridSearchCV
grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid, cv=10, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train)

# En iyi parametreler
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# En iyi model ile tahmin
best_classifier = grid_search.best_estimator_
y_pred = best_classifier.predict(X_test)

# Tahmin ve gerçek değerleri birleştirip yazdırma
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))

# Performans metrikleri
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)

# Tüm parametreler ve özellikler
print("\nBest Estimator Parameters:")
print(best_classifier.get_params())

# GridSearchCV sonuçları
print("\nGrid Search CV Results:")
print(grid_search.cv_results_)

```

Feature Engineering


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
from sklearn.preprocessing import StandardScaler

# Feature scaling ve KNeighborsClassifier pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures()),
    ('knn', KNeighborsClassifier())
])

# GridSearchCV için parametreler
param_grid = {
    'poly__degree': [1, 2, 3],
    'knn__n_neighbors': [3, 5, 7, 9, 11],
    'knn__metric': ['minkowski', 'euclidean', 'manhattan'],
    'knn__p': [1, 2]
}

# GridSearchCV
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=10, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train)

# En iyi parametreler
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# En iyi model ile tahmin
best_classifier = grid_search.best_estimator_
y_pred = best_classifier.predict(X_test)

# Tahmin ve gerçek değerleri birleştirip yazdırma
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))

# Performans metrikleri
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

Ensemble Methods

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Base estimators
estimators = [
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('tree', DecisionTreeClassifier(max_depth=3)),
    ('svc', SVC(kernel='rbf', probability=True))
]

# StackingClassifier
classifier = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
classifier.fit(X_train, y_train)

# Tahmin
y_pred = classifier.predict(X_test)

# Tahmin ve gerçek değerleri birleştirip yazdırma
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))

# Performans metrikleri
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

```python
# Sonuçları görselleştirin
leaderboard = predictor.leaderboard(train_data)
print(leaderboard)

# En iyi modelin adını alın
best_model_name = predictor.leaderboard(train_data).iloc[0, 0]
print("En iyi model:", best_model_name)
best_model = predictor._trainer.load_model(best_model_name)
print("En iyi modelin hiperparametreleri:", best_model.params)
# Scikit-learn KNeighborsClassifier modelini inceleyin
knn_model = best_model.model

# Modelin detaylarını yazdırın
print(knn_model)
print("Hiperparametreler:", knn_model.get_params())
# WeightedEnsemble_L2 modelinin içindeki modelleri öğrenin
ensemble_model_name = 'WeightedEnsemble_L2'
ensemble_model = predictor._trainer.load_model(ensemble_model_name)

# Ensemble modeli içinde kullanılan alt modellerin isimleri
base_model_names = ensemble_model.base_model_names
print("WeightedEnsemble_L2 modeli içinde kullanılan alt modeller:", base_model_names)
```
